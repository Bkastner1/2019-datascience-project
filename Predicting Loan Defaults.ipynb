```python
from scipy.io import arff
import pandas as pd

data = pd.DataFrame(arff.loadarff('german_credit_fruad.arff.txt')[0])
```

```python
# Cleaning Data

for column in data.columns:
    i = 0
    for item in data[column]:
        if isinstance(item, bytes) == True:
                fixed = item.decode("utf-8")
                data[column][i] = fixed
                print(i)
                
        i += 1
```
